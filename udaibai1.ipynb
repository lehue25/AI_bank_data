```python

import pandas as pd
df = pd.read_csv('/content/bank-data_xuly.csv')
```

```python
df.info()
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
attributes = ['age','sex','region','income','married','children','car',
            'save_act','current_act','mortgage','pep']
mean_values = df[attributes].mean()
plt.figure(figsize=(10,5))
mean_values.plot(kind='bar', color='skyblue')
plt.title('gia tri trung binh cua cac thuoc tinh trong bankdata')
plt.xlabel('Thuoc tinh')
plt.ylabel('Gia tri trung binh')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
attributes = ['age','sex','region','income','married','children','car',
            'save_act','current_act','mortgage','pep']
mean_values = df[attributes].mean()
plt.figure(figsize=(8,6))
plt.boxplot(df[attributes].values, labels = attributes)
plt.title('phan phoi cac thuoc tinh trong bankdata')
plt.xlabel('Thuoc tinh')
plt.ylabel('Gia tri')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
x_attribute = 'income'
y_attribute = 'age'
plt.figure(figsize=(8,6))
plt.scatter(df[x_attribute], df[y_attribute], color ='blue',alpha=0.5)
plt.title('Moi tuong quan giua thu nhap va vung mien ')
plt.xlabel('income')
plt.ylabel('age')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.hist(df['income'], bins=20, color='blue', edgecolor = 'black')
plt.title('Biểu đồ thu nhập Kh trong bankdata')
plt.xlabel('Income')
plt.ylabel('Tần suất')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = df['income']
y = df['age']
plt.figure(figsize=(8,6))
plt.plot(x,y, marker='o', linestyle="-")
plt.title('Mối quan hệ giữa độ tuổi và income')
plt.xlabel('income')
plt.ylabel('age')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
current_act_counts = df['current_act'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(current_act_counts, labels = current_act_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Phân phối giới tính')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
selected_columns = ['age','sex','region','income','married','children','car',
            'save_act','current_act','mortgage','pep']
selected_data = df[selected_columns]
pd.plotting.scatter_matrix(selected_data, figsize = (15,15))
plt.show()
```

```python
import matplotlib.pyplot as plt
correlation_matrix = df.corr()
plt.figure(figsize=(10,8))
plt.imshow(correlation_matrix, cmap = 'coolwarm', interpolation = 'nearest')
plt.colorbar(label='Correlation')
plt.title('Đồ tị tương quan của các thuộc tính trong bankdata')
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
for i in range(len(correlation_matrix)):
  for j in range(len(correlation_matrix)):
    plt.text(j,i,"{:.2f}".format(correlation_matrix.iloc[i,j]),
      ha='center', va='center', color='black')
plt.show()
```
